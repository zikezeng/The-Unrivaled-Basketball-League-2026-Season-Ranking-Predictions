```python
import pandas as pd

df_difference=pd.read_csv('key_metric_performance_difference.csv', header=None, names=['metric', 'value'])
df_difference['abs_value']=df_difference['value'].abs()
total_difference=df_difference['abs_value'].sum()
df_difference['weight']=df_difference['abs_value']/total_difference
print(df_difference)

formula=''
for i in range(len(df_difference)):
    metric=df_difference.loc[i, 'metric']
    weight=df_difference.loc[i, 'weight']
    if metric=='TO':
        part='-'+metric+'*'+str(round(weight, 2))
    else:
        part='+'+metric+'*'+str(round(weight, 2))
    formula=formula+part
if formula.startswith('+'):
    weight_formula=formula[3:]
print(weight_formula)
```
