```python
import requests
import pandas as pd

url = (
    "https://stats.wnba.com/stats/leagueLeaders"
    "?LeagueID=10"
    "&PerMode=PerGame"
    "&Scope=S"
    "&Season=2025"
    "&SeasonType=Regular+Season"
    "&StatCategory=PTS"
)

headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)",
    "Accept": "application/json, text/plain, */*",
    "Referer": "https://stats.wnba.com/leaders/",
    "Origin": "https://stats.wnba.com"
}

# data request
response = requests.get(url, headers=headers)
response.raise_for_status()
data = response.json()

# leagueLeaders uses resultSet (singular)
headers_cols = data["resultSet"]["headers"]
rows = data["resultSet"]["rowSet"]

# DataFrame
df = pd.DataFrame(rows, columns=headers_cols)

# Keep only the needed columns
cols_needed = ["PLAYER", "PTS", "STL", "AST", "TOV"]
df_small = df[cols_needed]

# save as new csv file
df_small.to_csv("../data/processed/wnba_2025_cleaned.csv", index=False)

print(df_small.head())



```
