```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('sorted_performance_difference.csv', header=None)
df.columns=['metrics', 'performance_difference']
plt.bar(df['metrics'], df['performance_difference'])
plt.title('Performance Difference by Metrics')
plt.xlabel('Metrics')
plt.ylabel('Performance Difference')
plt.show()
```
