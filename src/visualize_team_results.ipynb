```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("../data/processed/club_average_scores.csv")

# Sort by score 
df_sorted = df.sort_values("AVG_SCORE", ascending=False)

# Plot
plt.figure(figsize=(8,5))
plt.plot(df_sorted["club"], df_sorted["AVG_SCORE"], marker="o")

plt.title("Club Average Score")
plt.xlabel("Club")
plt.ylabel("Score")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
