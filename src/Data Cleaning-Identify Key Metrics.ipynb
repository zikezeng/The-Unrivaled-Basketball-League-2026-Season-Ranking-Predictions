```python
import pandas as pd
df_team=pd.read_csv('../data/raw/unrivaled_2025_team_stats.csv')

#Get the team name of the top 3 teams
rank_2025=df_team.sort_values('RK')
df_top3_only=rank_2025.head(3)
top3_teams_2025=df_top3_only['TEAM']
#Get the correspanding statistics of the top 3 teams
df_top3=df_team[df_team['TEAM'].isin(top3_teams_2025)]
#Select the metrics used for calculation
metrics=['PTS', 'OR', 'DR', 'REB', 'AST', 'STL', 'BLK', 'TO']
#Calculate the mean of each metric across the top 3 teams
top3_mean=df_top3[metrics].mean()
#Calculate the mean of each metric across the entire league
league_2025_mean=df_team[metrics].mean()
#Calculate the performance difference between the top 3 teams averages and the league's averages for each metric
difference={}
for metric in metrics:
    if metric == 'TO':
        difference[metric]=league_2025_mean[metric]-top3_mean[metric]
    else:
        difference[metric]=top3_mean[metric]-league_2025_mean[metric]

difference_series=pd.Series(difference)
sorted_difference=difference_series.sort_values(ascending=False)
sorted_difference.to_csv('sorted_performance_difference.csv', header=False)
print(sorted_difference)


key_metrics=['PTS', 'STL', 'AST', 'TO']
filtered_performance_difference=sorted_difference[key_metrics]
filtered_performance_difference.to_csv('key_metric_performance_difference.csv', header=False)

columns_to_keep=['TEAM']+key_metrics
df_team_key_metrics=df_team[columns_to_keep]
df_team_key_metrics.to_csv('unrivaled_2025_team_stats_processed.csv', index=False)
```
